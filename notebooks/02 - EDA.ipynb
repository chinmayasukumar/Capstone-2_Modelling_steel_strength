##  1.0 Importing data and packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
steel = pd.read_csv('/Users/chinmayasukumar/Documents/Springboard/Capstone #2/data/raw/steel_clean.csv')
steel = steel.drop(columns=['Unnamed: 0'])
```

## 3.0 Looking for correlations

```python
sns.pairplot(steel)
plt.show()
```

```python
sns.heatmap(steel.corr()[['yield', 'tensile', 'elongation', 'red_area']], cmap='coolwarm', annot=True)
```

#### Manganese, Nickel, Vanadium and Molybnenum seem to be the alloying elements that affect the strength of the steel the greatest

```python
fig, ax = plt.subplots(1,2)

ax[0].scatter(steel['mn'], steel['yield'], alpha=0.5)
ax[0].set_title('Manganese')
ax[0].set_xlabel('wt%')
ax[0].set_ylabel('Yield strength (MPa)')

ax[1].scatter(steel['ni'], steel['yield'], alpha=0.5)
ax[1].set_title('Nickel')
ax[1].set_xlabel('wt%')
ax[1].set_ylabel('Yield strength (MPa)')

plt.subplots_adjust(left=1, right=3)

plt.show()
```

```python
fig, ax = plt.subplots(1,2)

ax[0].scatter(steel['v'], steel['yield'], alpha=0.5)
ax[0].set_title('Vanadium')
ax[0].set_xlabel('wt%')
ax[0].set_ylabel('Yield strength (MPa)')

ax[1].scatter(steel['mo'], steel['yield'], alpha=0.5)
ax[1].set_title('Molybdenum')
ax[1].set_xlabel('wt%')
ax[1].set_ylabel('Yield strength (MPa)')

plt.subplots_adjust(left=1, right=3)

plt.show()
```

#### None of the alloying element are strongly correlated to the strength of the steel. Further investigation is needed.
