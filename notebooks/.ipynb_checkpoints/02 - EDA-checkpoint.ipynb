## 3.0 EDA

```python
def hist_plotter(xs, titles, x_labels):
    fig, ax = plt.subplots(8, 2, figsize=(14,30), constrained_layout=True, dpi=200)
    fig.suptitle('Distribution of features')
    
    index = 0
    
    for i in range(8):
        for j in range(2):
            if (i == 7 and j == 1):
                plt.show()
            else:  
                sns.histplot(steel[xs[index]],ax=ax[i,j])
                ax[i,j].set_title(titles[index])
                ax[i,j].set_xlabel(x_labels[index])
                index += 1
```

```python
xs = ['c','si', 'mn', 'p', 's', 'ni', 'cr', 'mo', 'cu', 'v', 'al', 'n', 'nb+ta', 'temp', 'yield',]
titles = ['Carbon', 'Silicon', 'Manganese', 'Phosphorous', 'Sulfur', 'Nickel', 'Chromium', 'Molybdenum', 'Copper',\
          'Vanadium', 'Aluminum', 'Nickel', 'Niobium + Tantalum', 'Temperature', 'Yield Strength']
x_labels = ['wt%' for i in range(13)] + ['degC', 'MPa']
```

```python
hist_plotter(xs, titles, x_labels)
```

#### Yield strength will be chosen as the target variable since it usually most relevant in engineering applications

```python
def plotter(x, x_label, title, i_x, i_y):
    hue_order = [27, 100, 150, 200, 250, 300, 350, 400, 450]
    ax[i_x, i_y].scatter(steel[x], steel['yield'], alpha=0.6)
    ax[i_x, i_y].set_title(title)
    ax[i_x, i_y].set_xlabel(x_label)
    ax[i_x, i_y].set_ylabel('Yield strength (MPa)')
```

```python
fig, ax = plt.subplots(3,2,figsize=(17,15), constrained_layout=False)

plotter('mn', 'wt%', 'Manganese', 0, 0)
plotter('ni', 'wt%', 'Nickel', 0, 1)
plotter('v', 'wt%', 'Vanadium', 1, 0)
plotter('mo', 'wt%', 'Molybdenum', 1, 1)
plotter('c', 'wt%', 'Carbon', 2, 0)
plotter('temp', 'Temperature (degC)', 'Temperature', 2, 1)

plt.show()
```

#### There are some correlations with a few elements and the yield strength
